```python
import matplotlib.pyplot as plt

#activation functions

arrx = []
arrybin = []
arrybip = []
arryplin = []
for x in range (-100,100):
  arrx.append(x/10)
for val in arrx:
  if val>=0:
    arrybin.append(1)
    arrybip.append(1)
  else:
    arrybin.append(0)
    arrybip.append(-1)
for val in arrx:
  if val>=2:
    arryplin.append(1)
  elif val>-2:
    arryplin.append((val/4)+0.5)
  else:
    arryplin.append(0)

#plt.plot(arrx, arrybin, label='binary step')
#plt.plot(arrx, arrybip, label='bipolar step')
#plt.plot(arrx, arrx, label = 'linear')
plt.plot(arrx, arryplin, label = 'piece-wise linear')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#piecewise linear

arrx = []
arryplin = []
for x in range (-100,100):
  arrx.append(x/10)
for val in arrx:
  if val>=5:
    arryplin.append(1)
  elif val>-5:
    arryplin.append(val/5)
  else:
    arryplin.append(-1)

plt.plot(arrx, arryplin, label = 'piece-wise linear')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

#sigmoid activation function

arrx = []
arry1 = []
arry2 = []
arry3 = []

for x in range (-100,100):
  arrx.append(x/10)

for val in arrx:
  y1 = 1/(1+((math.e)**(-(val))))
  arry1.append(y1)
  y2 = 1/(1+((math.e)**(-2*(val))))
  arry2.append(y2)
  y3 = 1/(1+((math.e)**(-0.5*(val))))
  arry3.append(y3)

plt.plot(arrx, arry1, label = 'slope = 1')
plt.plot(arrx, arry2, label = 'slope = 2')
plt.plot(arrx, arry3, label = 'slope = 0.5')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

#hyperbolic tangent function

arrx = []
arry1 = []
arry2 = []
arry3 = []

for x in range (-100,100):
  arrx.append(x/10)

for val in arrx:
  y1 = (((math.e)**val)-(math.e)**(-val))/(((math.e)**val)+(math.e)**(-val))
  arry1.append(y1)
  y2 = ((math.e)**(2*val)-(math.e)**(-2*val))/((math.e)**(2*val)+(math.e)**(-2*val))
  arry2.append(y2)
  y3 = ((math.e)**(0.5*val)-(math.e)**(-0.5*val))/((math.e)**(0.5*val)+(math.e)**(-0.5*val))
  arry3.append(y3)

plt.plot(arrx, arry1, label = 'slope = 1', color = 'violet')
plt.plot(arrx, arry2, label = 'slope = 2', color = 'pink')
plt.plot(arrx, arry3, label = 'slope = 0.5', color = 'purple')
plt.legend()
plt.show()
```
